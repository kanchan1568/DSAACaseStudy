**DATA CLEANING**

```python
# Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Check dataset information
print("\nDataset Info:")
print(df.info())

# Remove unnecessary column
if 'Unnamed: 22' in df.columns:
    df.drop('Unnamed: 22', axis=1, inplace=True)

# Remove extra spaces from text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()

print("\nColumns after cleaning:")
print(df.columns)
```

**HANDLING MISSING VALUES**

```python
print("\nMissing Values Before Treatment:")
print(df.isnull().sum())

# Numerical columns -> fill with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns -> fill with mode
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing Values After Treatment:")
print(df.isnull().sum())
```

**DUPLICATE REMOVAL**

```python
duplicates = df.duplicated().sum()

print("\nNumber of Duplicate Rows:", duplicates)

df.drop_duplicates(inplace=True)

print("Shape After Removing Duplicates:", df.shape)
```

**DATA TYPE CONVERSION**

```python
# Convert Date column into datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Convert postal code to string
df['ship-postal-code'] = df['ship-postal-code'].astype(str)

# Convert B2B column to boolean
df['B2B'] = df['B2B'].astype(bool)

print("\nUpdated Data Types:")
print(df.dtypes)
```

**SAVE PREPROCESSED DATA**

```python
df.to_csv("Amazon_Sale_Report_Preprocessed.csv", index=False)

print("\nPreprocessing Completed Successfully!")
print("Cleaned file saved as Amazon_Sale_Report_Preprocessed.csv")
```

4.1 Data Cleaning

Removed unnecessary spaces from column names and text values.
Dropped the irrelevant column Unnamed: 22.
Inspected dataset structure and attributes.

4.2 Handling Missing Values

Missing numerical values were replaced using the median.
Missing categorical values were replaced using the mode.
This ensured no significant information was lost.

4.3 Duplicate Removal

Duplicate records were identified using duplicated().
Duplicate rows were removed using drop_duplicates().

4.4 Data Type Conversion

Converted Date to datetime format.
Converted ship-postal-code to string format.
Converted B2B to Boolean format for easier analysis.
